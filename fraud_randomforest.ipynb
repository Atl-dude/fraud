```python
# Detecting Fraud

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data = pd.read_csv("/kaggle/input/fraud/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/fraud/test.csv")
test_data.head()
```

```python
ok_MasterCard = train_data.loc[train_data.Card == 'MasterCard']["ok"]
rate_MasterCard = sum(ok_MasterCard)/len(ok_MasterCard)

print("% of MasterCard that Failed:", rate_MasterCard)
```

```python
ok_Visa = train_data.loc[train_data.Card == 'Visa']["ok"]
rate_Visa = sum(ok_Visa)/len(ok_Visa)

print("% of Visa that Failed:", rate_Visa)
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Passed"]

features = ["Zip", "Card", "Store_Type", "Brick_or_Online"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'CustomerId': test_data.CustomerId, 'Failed': predictions})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
